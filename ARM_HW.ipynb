```python
#Setting up the environment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
# Now let's import the dataset and see what we're working with. Please download the store_data.csv into your WORKING FOLDER
# and read it in as a DataFrame. Name it store_data

# A small yet important detail whenever you import data -- check whether the first row contains header or not
# In this case, there is no header -- check the details of the "read_csv" method and see what parameter you need to 
# tweak to reflect that

store_data =  pd.read_csv('store_data.csv', header=None)
store_data.head()
```

```python
#Let's check the data set
#Look at the first five rows. If you imported properly, you should see something like below:
```

```python
# Here is the catch: the Apriori library we are going to use requires our dataset to be in the form of a LIST of lists (don't ask me why) 
# We need to view the whole dataset as a big list, and each transaction in the dataset an inner list within the outer big list. 
#Currently we have data in the form of a pandas dataframe. To convert our pandas dataframe into a list of lists, execute the following script:
records = []
for i in range(0, 7501):
    records.append([str(store_data.values[i,j]) for j in range(0, 20)])
```

```python
# Time to run the apriori! The apriori class requires some parameter values to work. 
# The first parameter is the list of list that you want to extract rules from. 
# The second parameter is the min_support parameter. 
# Next, the min_confidence parameter filters those rules that have confidence greater than the confidence threshold specified by the parameter. 
# Similarly, the min_lift parameter specifies the minimum lift value for the short listed rules. 
# Finally, the min_length parameter specifies the minimum number of items that you want in your rules.
# Play with these parameters. Choose one that you feel give you interesting rules. 
# Depending on your parameters, your results might be different from mine: 
association_rules = apriori(records, min_support= 0.0045 , min_confidence= 0.2 , min_lift= 3, min_length= 2)
association_results = list(association_rules)
print(len(association_results))
```

```python
# Time to print the rules you collected! 
# Try command like the following, interpret your results
print(association_results[0])
```
